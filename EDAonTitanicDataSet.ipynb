```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To convert .csv file to a dataframe
df= pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
#To see 5 rows of the whole data as a dataframe
df.head()
```

```python
#Since PassengerId has nothing important to do with any of the existing data iloc(integer location) converts the df to have all the rows but just the columns from the 1st one
df.iloc[:,1:]
```

```python
#To get overall info of the dataframe
df.info()
```

```python
#To find correlation of all the numeric columns with each other
df.corr(numeric_only=True)
```

```python
#To get the mathematical information about every columns
df.describe()
```

```python
#To find total null values in each column
df.isnull().sum()
```

```python
#To get a numerical representation of the Survived column
sns.countplot(x=df['Survived'])
```

```python
#To get percentage depiction of how many people survived
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
#To get an axial representation of how columns Fare,Age,Survived,Pclass and Sex are related to each other
sns.scatterplot(x=df['Age'],y=df['Fare'],hue=df['Survived'],style=df['Sex'],size=df['Pclass'])
```

```python
#To get bar representation between the relation of Age Survived and Sex
sns.barplot(x=df['Survived'],y=df['Age'],hue=df['Sex'])
```

```python
#To get a density point of view on the relation of Age and Survived column
sns.kdeplot(x=df['Age'],hue=df['Survived'])
```

```python
#To get an idea to how the columns are correlated to each other on basis of colur.In the rage of -1 to 1 more the value more correlation
corr=df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
```
